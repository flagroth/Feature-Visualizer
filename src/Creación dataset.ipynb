```python
import pandas as pd
```

```python
attributes = pd.read_csv("../data/PlayerAttributeData.csv")
personal = pd.read_csv("../data/PlayerPersonalData.csv")
```

```python
dataset = pd.concat([personal, attributes], axis=1)
dataset.columns
```

```python
dataset = dataset.drop(['Unnamed: 0','Unnamed: 0.1','Photo','Flag', 
                        'Nationality','Club Logo','Value','Wage',
                        'Club','Special'], axis = 1)
dataset.columns
```

```python
dataset.to_csv("../data/dataset_attributes.csv", index = False)
```
